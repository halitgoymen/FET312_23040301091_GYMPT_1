1-Gerekli KÃ¼tÃ¼phaneleri indiriyorum.

```python
import torch
import sys
import requests
from PIL import Image
from io import BytesIO
from transformers import (
    Qwen2VLForConditionalGeneration, 
    AutoProcessor, 
    TrainingArguments, 
    Trainer,
    BitsAndBytesConfig,
    DataCollatorForSeq2Seq  
)
import bitsandbytes as bnb
from peft import (
    LoraConfig, 
    get_peft_model, 
    prepare_model_for_kbit_training,
    PeftModel
)
from datasets import load_dataset
import time
import pandas as pd
```

2-GPU kullanmamÄ±z gerektiÄŸi iÃ§in burada GPU mu yoksa CPU mu kullanÄ±yoruz diye kontrol ediyorum.

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Ã‡alÄ±ÅŸma OrtamÄ±: {device.upper()}")
if device == "cuda":
    print(f"GPU Modeli: {torch.cuda.get_device_name(0)}")
    print(f"VRAM Bilgisi: {torch.cuda.get_device_properties(0).total_memory / 1e9:.2f} GB")
else:
    print("UYARI: CPU kullanÄ±yorsunuz. Model Ã§ok yavaÅŸ Ã§alÄ±ÅŸacaktÄ±r!")
```

3-KullanacaÄŸÄ±mÄ±z model olan Qwen2-VL-2B'yi burada indiriyorum.

```python
model_name = "Qwen/Qwen2-VL-2B-Instruct"

model = Qwen2VLForConditionalGeneration.from_pretrained(
    model_name,
    torch_dtype=torch.bfloat16,
    attn_implementation="sdpa"
).to("cuda") 

processor = AutoProcessor.from_pretrained(model_name)
processor.tokenizer.pad_token = processor.tokenizer.eos_token

model.gradient_checkpointing_enable()
model.enable_input_require_grads()
```

4-Modelin eÄŸitilmemiÅŸ hail iÃ§in basit prompt test yapÄ±yorum.

```python
question = "Sen yardÄ±mcÄ± bir aÅŸÃ§Ä±sÄ±n. Sadece Ã¶ÄŸle yemeÄŸi iÃ§in, toplamÄ± yaklaÅŸÄ±k 500 kalori olan saÄŸlÄ±klÄ± ve tek bir tabak yemek tarifi veya menÃ¼sÃ¼ Ã¶ner."
text_prompt = processor.apply_chat_template(
    [{"role": "user", "content": question}],
    tokenize=False,
    add_generation_prompt=True
)

inputs = processor(
    text=[text_prompt],
    images=None,
    return_tensors="pt",
    padding=True
).to(model.device)

output = model.generate(
    **inputs,
    max_new_tokens=256,
    do_sample=True,
    temperature=0.7,
    top_p=0.9,
    repetition_penalty=1.1
)

answer = processor.batch_decode(output, skip_special_tokens=True)[0]
print(answer.split("assistant")[-1].strip())
```

5-Modelin eÄŸitilmemiÅŸ hail iÃ§in basit gÃ¶rsel test yapÄ±yorum.

```python
url = "https://images.unsplash.com/photo-1546069901-ba9599a7e63c?ixlib=rb-4.0.3&auto=format&fit=crop&w=800&q=80" 
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    image = Image.open(BytesIO(response.content))
else:
    raise Exception(f"Resim indirilemedi! Kod: {response.status_code}")

prompt = """
Sen uzman bir diyetisyensin. Bu fotoÄŸraftaki yiyecekleri analiz et.
Sadece tespit ettiÄŸin yiyecekleri ve USDA standartlarÄ±na gÃ¶re yaklaÅŸÄ±k kalori deÄŸerlerini listele.
Asla kendini tekrar etme. Ã‡Ä±ktÄ± formatÄ± kesinlikle ÅŸÃ¶yle olsun:

- [Yiyecek AdÄ±]: [Kalori DeÄŸeri] kcal
...
TOPLAM: [Toplam Kalori] kcal
"""

conversation = [
    {
        "role": "user",
        "content": [
            {"type": "image", "image": image},
            {"type": "text", "text": prompt},
        ],
    }
]

text_prompt = processor.apply_chat_template(conversation, tokenize=False, add_generation_prompt=True)
inputs = processor(text=[text_prompt], images=[image], return_tensors="pt", padding=True).to(model.device)
with torch.no_grad():
    output_ids = model.generate(
        **inputs,
        max_new_tokens=256,
        do_sample=True,        
        temperature=0.2,       
        top_p=0.9,
        repetition_penalty=1.2 # TekrarÄ± Ã¶nle
    )

generated_ids = output_ids[0][len(inputs.input_ids[0]):] 
response_text = processor.decode(generated_ids, skip_special_tokens=True)

print("-" * 30)
print(response_text.strip())
print("-" * 30)
```

6-Kendi hazÄ±rladÄ±ÄŸÄ±mÄ±z datasett ile eÄŸitmek iÃ§in datasetti Ã§aÄŸÄ±rÄ±p tokenizasyon iÅŸlemini gerÃ§ekleÅŸtiriyorum.

```python
dataset = load_dataset("json", data_files="train_data_chat.jsonl")

def tokenize_function(example):
    text = processor.apply_chat_template(
        example["messages"], 
        tokenize=False, 
        add_generation_prompt=False
    )
    tokens = processor.tokenizer(
        text,
        truncation=True,
        max_length=1024,
        padding="max_length",
    )
    
    
    labels = tokens["input_ids"].copy()
    
    pad_token_id = processor.tokenizer.pad_token_id
    for i, label in enumerate(labels):
        if label == pad_token_id:
            labels[i] = -100
            
    tokens["labels"] = labels
    return tokens

tokenized_dataset = dataset.map(tokenize_function, remove_columns=dataset["train"].column_names)
print("Veri iÅŸlendi ve padding maskelendi.")
```

7-Low-Rank Adaptation ile modeli kendi ekran kartÄ±ma uygun hale getiriyorum.

```python
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"], 
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
```

8-EÄŸitim iÃ§in gerekli olan epoch, learning rate gibi deÄŸerleri burada belirleyip eÄŸitimi baÅŸlatÄ±yorum.

```python
training_args = TrainingArguments(
    output_dir="./qwen2vl_2b_diet_bot",
    per_device_train_batch_size=1,
    gradient_accumulation_steps=16,
    num_train_epochs=10,
    learning_rate=2e-4,
    bf16=True,
    logging_steps=5,
    save_strategy="epoch",
    save_total_limit=2,
    report_to="none",
    remove_unused_columns=False,
    optim="paged_adamw_8bit", 
    dataloader_pin_memory=False
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    data_collator=DataCollatorForSeq2Seq(processor.tokenizer, padding=True, pad_to_multiple_of=8)
)

print("EÄŸitim BaÅŸlÄ±yor...")
trainer.train()
```

9-EÄŸittiÄŸim ilk ilkel halini burada test ediyorum.

```python
model.eval()
soru = "Hocam kilo vermek istiyorum ama tatlÄ± krizim tutuyor, ne yapmalÄ±yÄ±m?"
messages = [
    {
        "role": "system", 
        "content": "Sen yardÄ±msever ve bilgili bir diyetisyensin. Sorulara kÄ±sa, net ve bilimsel Ã¶nerilerle cevap ver."
    },
    {"role": "user", "content": soru}
]
text = processor.apply_chat_template(
    messages, 
    tokenize=False, 
    add_generation_prompt=True
)
inputs = processor(
    text=[text], 
    return_tensors="pt", 
    padding=True
).to("cuda")
generated_ids = model.generate(
    **inputs,
    max_new_tokens=230,     
    do_sample=True,         
    temperature=0.2,       
    top_p=0.85,             
    top_k=50,               
    repetition_penalty=1.2  
)

generated_ids = [
    output_ids[len(input_ids):] for input_ids, output_ids in zip(inputs.input_ids, generated_ids)
]
response = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

print("-" * 30)
print(f"SORU: {soru}")
print(f"DÄ°YETÄ°SYEN CEVABI:\n{response}")
print("-" * 30)
```

10-5 Test veri setim ile modeli deniyorum.

```python

test_questions = [
    "Hocam su iÃ§meyi unutuyorum, ne Ã¶nerirsiniz?",
    "Protein tozu zararlÄ± mÄ±?",
    "AkÅŸam 8'den sonra meyve yenir mi?",
    "Vegan besleniyorum, protein ihtiyacÄ±mÄ± nasÄ±l karÅŸÄ±larÄ±m?",
    "GÃ¶bek yaÄŸlarÄ±nÄ± eritmek iÃ§in Ã¶zel bir kÃ¼r var mÄ±?"
]

results = []

print("DOÄRULUK TESTÄ° BAÅLIYOR.\n")


generation_kwargs = {
    "max_new_tokens": 150,
    "do_sample": True,
    "temperature": 0.2,      
    "top_p": 0.9,
    "repetition_penalty": 1.2
}


model.eval() 

for i, question in enumerate(test_questions):
    print(f"Soru {i+1}/{len(test_questions)} iÅŸleniyor: {question}")
    
    
    start_time = time.time()
    
    
    messages = [
        {"role": "system", "content": "Sen yardÄ±msever ve bilgili bir diyetisyensin. KÄ±sa ve net cevap ver."},
        {"role": "user", "content": question}
    ]
    
    text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = processor(text=[text], return_tensors="pt", padding=True).to("cuda")
    
   
    generated_ids = model.generate(**inputs, **generation_kwargs)
    
   
    generated_ids = [output_ids[len(input_ids):] for input_ids, output_ids in zip(inputs.input_ids, generated_ids)]
    response = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
    
    
    end_time = time.time()
    duration = end_time - start_time
    
    
    results.append({
        "Soru": question,
        "Model CevabÄ±": response,
        "SÃ¼re (sn)": round(duration, 2)
    })


print("\n" + "="*50)
print("TEST SONUÃ‡ RAPORU")
print("="*50)

df_results = pd.DataFrame(results)

for index, row in df_results.iterrows():
    print(f"\nâ“ SORU: {row['Soru']}")
    print(f"ğŸ¤– CEVAP: {row['Model CevabÄ±']}")
    print(f"â±ï¸ SÃœRE: {row['SÃ¼re (sn)']} saniye")
    print("-" * 30)
```
